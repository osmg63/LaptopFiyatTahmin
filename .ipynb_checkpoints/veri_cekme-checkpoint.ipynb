```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import win_unicode_console
win_unicode_console.enable()
```

```python

url_list=[]
prices_list=[]
propTitles=[]
propValues=[]
propMarka=[]
for i in range(1,50):
    print(i)
    url ="https://www.trendyol.com/oyuncu-dizustu-bilgisayari-x-c106084?pi="+str(i)
    r=requests.get(url)
    source=BeautifulSoup(r.content,"lxml")
    
    urls = source.find_all("div",attrs={"class":"p-card-wrppr with-campaign-view"})
    
    for url in urls:
        url_laptop="https://www.trendyol.com/"+url.a.get("href")
        url_list.append(url_laptop)
        #print(url_laptop)
        
        r_laptop=requests.get(url_laptop)
        source_laptop = BeautifulSoup(r_laptop.content, "lxml", from_encoding='utf-8')  # from_encoding ekleyin
        

        properties = source_laptop.find_all("li", class_="detail-attr-item")
        for prop in properties:
            prop_title = prop.find("span").text
            prop_value = prop.find("b").text
            propTitles.append(prop_title)
            propValues.append(prop_value)
          #  print(prop_title+"  "+prop_value)
       
    prices=source.find_all("div",attrs={"class":"prc-box-dscntd"})
    for price in prices:
        prices_list.append(price.text)
       # print(price.text)
    marka=source.find_all("span",attrs={"class":"prdct-desc-cntnr-ttl"})
    for mark in marka:
        propMarka.append(mark.text)
        #print(mark.text)
        
        
print(str(len(propMarka))+" Adet marka bulundu")
print(str(len(url_list))+" Adet link bulundu")
print(str(len(prices_list))+" Adet fiyat bulundu")
print(str(len(propTitles))+" Adet özellik başlığı bulundu")
print(str(len(propValues))+" Adet özellik verisi bulundu")
```

```python
df_urls=pd.DataFrame()
df_urls["urls"]=url_list
df_urls["prices"]=prices_list
df_urls["marka"]=propMarka
```

```python
laptops = len(url_list)
```

```python
columns=np.array(propTitles)
columns=np.unique(columns)
```

```python
okulaDonus=[]
df=pd.DataFrame(columns=columns)
df["url"]=url_list
df["price"]=prices_list
df["marka"]=propMarka
```

```python
df.to_csv("deneme.csv", index=0)
```

```python
for i in range(0,laptops):
    print(i)
    url=df['url'].loc[i]
    r=requests.get(url)
    source=BeautifulSoup(r.content, "lxml", from_encoding='utf-8')
    
    properties = source.find_all("li", class_="detail-attr-item")
    for prop in properties:
        prop_title = prop.find("span").text
        prop_value = prop.find("b").text
        #print(prop_title+":"+prop_value)
        df[prop_title.encode('utf-8').decode('utf-8')].loc[i] = prop_value
```
